```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,10)
y =  x**2
```

##  function

```python
plt.plot(x,y,'r-|')
plt.xlabel("X Lable")
plt.ylabel("Y Lable")
plt.title("Title labe")
plt.show()
```

# multiplot 

```python
plt.subplot(1,2,1)
plt.xlabel('X LABLE' )
plt.ylabel('Y LABLE')
plt.title('TITILE LABLE 1')

plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.title('TITILE LABLE 2')
```

# object oriented 

```python
fig = plt.figure()
axes = fig.add_axes([0.1 ,0.1 ,0.8,0.8])
axes.plot(x, y, 'b')
axes.plot(x,y)
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1 ,0.1 ,0.8,0.8])
axes2 = fig.add_axes([0.21 ,0.47 ,0.3,0.3])
axes1.plot(x, y, 'b')
axes1.plot(x,y)
axes1.set_xlabel('Axes1  X Label') # Notice the use of set_ to begin methods
axes1.set_ylabel('Axes1  y Label')
axes1.set_title('Axes1  Title')

axes2.plot(x, y, 'b')
axes2.plot(y,x,'r')
axes2.plot(y,x*2,'r')
axes2.plot(y,x*4,'r')

axes2.set_xlabel('Axes2  Label') # Notice the use of set_ to begin methods
axes2.set_ylabel('Axes2  Label')
axes2.set_title(' Axes2 Title')

```

## subplot

```python
fig ,axes = plt.subplots()
axes.plot(x,y)
```

```python
fig ,axes = plt.subplots(nrows=4,ncols=4  )
plt.tight_layout() # important
```

```python
fig ,axes = plt.subplots(nrows=2,ncols=1 ,figsize=(2,3),dpi=150 )
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
plt.tight_layout()
```

```python
fig.savefig("filename", dpi=200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2, label="x**2")
axes.plot(x,x**3, label="x**3")
axes.legend()
```

```python
fig , ax = plt.subplots(dpi=100)
ax.plot(x,x, color='blue',alpha=0.1)
ax.plot(x,x+1, color='#FF8C00')
ax.plot(x,x+2, color='#8B008B')
```

### line and marker style

```python
fig ,ax = plt.subplots(figsize=(12,6),dpi=100)
ax.plot(x,x-2 ,color='blue' , lw=0.2)
ax.plot(x,x-1 ,color='blue' , lw=0.3)
ax.plot(x,x ,color='blue' , lw=0.8)
ax.plot(x,x+1 ,color='blue' , lw=3)
#linestype options : .. ,-. three option

ax.plot(x,x+2 ,color='green' , lw=3,ls=':')
ax.plot(x,x+3 ,color='green' , lw=3,ls='--')
ax.plot(x,x+4 ,color='green' , lw=3 ,ls='-.')
#custom dash 

line ,=ax.plot(x,x+5,color='black',lw=2)
line.set_dashes([1,10,11,20])#format: line length, space length


# marker symble +,*,o,,,,1,2,3,....
ax.plot(x,x+6 ,color='red' , marker='o',lw=1,markersize=10)
ax.plot(x,x+7 ,color='red' , marker='*',lw=1,markersize=10)
ax.plot(x,x+8 ,color='red' , marker='+',lw=1,markersize=10)
ax.plot(x,x+9 ,color='red' , marker='1',lw=1,markersize=10)


# marker size and color 
ax.plot(x,x+10 ,color='darkblue' , marker='o',lw=1,markersize=1 )
ax.plot(x,x+11 ,color='darkblue' , marker='o',lw=1,markersize=3 )
ax.plot(x,x+12 ,color='darkblue' , marker='o',lw=1,markersize=6 )
ax.plot(x,x+13 ,color='darkblue' , marker='o',lw=1,markersize=6 ,markerfacecolor='red' )
ax.plot(x,x+13 ,color='darkblue' , marker='s',lw=1,markersize=10 ,markerfacecolor='blue' ,markeredgewidth=3,markeredgecolor='yellow' )

ax.plot(x,x+15 ,color='darkblue' , marker='o',lw=1,markersize=10 ,markerfacecolor='yellow' ,markeredgewidth=3,markeredgecolor='red' )
ax.plot(x,x+18 ,color='darkblue' , marker='+',lw=1,markersize=10 ,markerfacecolor='yellow' ,markeredgewidth=3,markeredgecolor='black' )
```

```python
fig.savefig("line_marker_style.png",dpi=300)
```

# plot range

```python
fig ,ax =plt.subplots(nrows=1,ncols=3,figsize=(12,4))
ax[0].plot(x,x**2,x,x**3)
ax[0].set_title("default")

ax[1].plot(x,x**2,x,x**3)

ax[1].axis("tight")
ax[1].set_title("tight axes")

ax[2].plot(x,x**2,x,x**3)
ax[2].set_ylim([0,60])
ax[2].set_xlim([2,5])



```
