# plot Type
* df.plot.area     
* df.plot.barh  
* df.plot.bar   
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter   
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
