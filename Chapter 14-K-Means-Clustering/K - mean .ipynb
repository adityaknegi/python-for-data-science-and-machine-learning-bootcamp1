```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
 from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0].shape # our sample with 4 blob 
data[1]
```

```python

fig =plt.figure(figsize=(12,6),dpi=200)

plt.scatter(x=data[0][:,1],y=data[0][:,0],marker='o',cmap='rainbow',c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=10)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f ,(ax1,ax2) = plt.subplots(1,2,sharex='all',figsize=(12,6))##Share both X and Y axes with all subplots
ax1.set_title('result')
ax2.set_title('Original')
ax2.scatter(x=data[0][:,1],y=data[0][:,0],marker='o',cmap='rainbow',c=data[1])
ax1.scatter(x=data[0][:,1],y=data[0][:,0],marker='o',cmap='rainbow',c=kmeans.labels_)
```
