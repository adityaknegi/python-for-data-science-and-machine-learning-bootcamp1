# Categorical Data plot

* barplot
* countplot
* boxplot
* violinplot
* striplot
* swimplot
* catoplot  // old name factorplot

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# barplot

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
sns.barplot(x='day',y='tip',data=tips)
```

```python
sns.barplot(x='tip',y='day',data=tips)
```

# countplot

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.countplot(x='time',data=tips)
```

# boxplot

```python

sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
import matplotlib.pyplot as plt
tips1=tips[tips['smoker']=='No']
tips2=tips[tips['smoker']=='Yes']




sns.boxplot(x='day',y='total_bill',data=tips1,hue='sex')
fig =plt.figure()
axes = fig.add_axes([0,0,1,1])
sns.boxplot(x='day',y='total_bill',data=tips2,hue='sex')
```

# voilinplot 

```python
sns.violinplot(x='sex', y='total_bill', data=tips,hue='day',palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex',palette='rainbow')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', split=True, palette='Set1')
```

# striplot

```python
sns.stripplot(x='day',y='total_bill',data=tips) # Jitter=True by default 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=False) # Jitter=True by default 
#but know many point are stack on each other 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex') # Jitter=True by default 
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex' , dodge=True) # Jitter=True by default 
```

# swarmplot

combination of stripplot and violan plot

drawback :  no scale when large dataset , more time consuming

```python
sns.swarmplot(x='sex' ,y='total_bill' , data=tips)
```

```python
sns.swarmplot(x='sex' ,y='total_bill' , data=tips, color='black')
sns.violinplot(x='sex', y='total_bill', data=tips)
```

# factorplot

```python
sns.catplot(x='day' ,y='total_bill',kind='bar',data=tips)
```
