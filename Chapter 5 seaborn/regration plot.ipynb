```python
import seaborn as sns 
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips )
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex' )
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex'  ,markers=['o','*'] ,scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex'  ,markers=['o','*'] ,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex'  ,markers=['o','*'] ,col='day',row='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex'  ,col='day' ,aspect=0.6 ,height=10)
```
